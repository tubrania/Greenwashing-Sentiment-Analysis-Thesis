# Sentiment Analysis in Reports

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from transformers import pipeline
```

### Read Sentences

```python
# Define the file path to the CSV file in the "data" directory
file_path = "../data/reports/sentences_from_reports.csv"

# Read the CSV file
df_s = pd.read_csv(file_path)


# Ensure the 'sentence' column is of string data type
df_s["sentence"] = df_s["sentence"].astype(str)
```

### Analyze Sentiment

```python
classifier = pipeline("text-classification", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")
```

```python
sent_list = df_s['sentence'].tolist()
```

```python
%%time

CNT = 100
LEN = len(sent_list)
result = []
for i in range(0, LEN, CNT):
    off_e = (i+CNT) if (i+CNT) < LEN else LEN
    
    print(f'sent_list[{i}:{off_e}]')
    # Max length of sentence: 512
    res = classifier(sent_list[i:off_e], truncation=True)
    
    result.extend(res)
        
print('==== End of jobs ====')
```

### Labels and Scores

```python
label = []
score = []
for dic in result:
    label.append(dic['label'])
    score.append(dic['score'])
df_s['label'] = label
df_s['score'] = score
```

```python
df_s['label'] = df_s['label'].str.title()
df_s['score'] = df_s['score'].round(2)
```

```python
df_s
```

```python
df_sentiment = df_s[['company','year','label','score']].copy()
```

```python
# Group the DataFrame 'df_sentiment' by 'company' and 'year' and count the occurrences of each 'label' within each group
df_comp_cnt = df_sentiment.groupby(['company', 'year'])['label'].apply(lambda x: x.value_counts()).unstack()

# Calculate the ratio of 'Positive' to 'Negative' counts within each group and round it to two decimal places
df_comp_cnt['Ratio'] = (df_comp_cnt['Positive'] / df_comp_cnt['Negative']).round(2)

# Display the resulting DataFrame 'df_comp_cnt' that now contains counts and the 'Ratio' column
df_comp_cnt
```

```python
# Reset the index of the DataFrame 'df_comp_cnt', making 'company' and 'year' columns regular columns
df_comp_cnt_flat = df_comp_cnt.reset_index()

# Display the resulting DataFrame 'df_comp_cnt_flat'
df_comp_cnt_flat
```

### Save the Result 

```python
df_comp_cnt_flat.to_csv('sentiments.csv', index=False)
```
